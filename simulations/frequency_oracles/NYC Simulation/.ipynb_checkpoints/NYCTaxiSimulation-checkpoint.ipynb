# 1) LDP Heatmap Analysis using Clustering + matplot/folium

## The code below uses Mini Batch K-Means clustering with 2000 clusters on 1,000,000 taxicab rides to be used as LDP Bins.

```python
test_data = pd.read_csv("data/train.csv")
geodata = np.column_stack((test_data["dropoff_latitude"], test_data["dropoff_longitude"]))
reduced_geodata = geodata[np.random.choice(geodata.shape[0], 100000, replace=False)]

clustering = MiniBatchKMeans(n_clusters=2000).fit(reduced_geodata)

plt.figure(figsize=(20, 20))
plt.axis([40.6, 40.9, -74.25, -73.6])

# plt.scatter(clustering.cluster_centers_[:,0], clustering.cluster_centers_[:, 1])
cm = plt.get_cmap("magma")
plt.scatter(reduced_geodata[:, 0], reduced_geodata[:, 1],  c=clustering.labels_, s=0.3, cmap="magma")
plt.show()

cluster_data = clustering.predict(geodata)
```

## The plot below shows all 2000 clusters, captures the shape of NYC relatively well

```python
cm = plt.get_cmap("magma")
plt.figure(figsize=(20, 20))
plt.axis([40.6, 40.9, -74.25, -73.6])
plt.scatter(clustering.cluster_centers_[:, 0], clustering.cluster_centers_[:, 1], c=range(0,2000), s=1, cmap="magma")
plt.show()
```

```python
from sklearn.preprocessing import normalize

counter = Counter(cluster_data)

centers = []
weights = []
max_freq = counter.most_common(1)[0][1]
print("Max freq cluster:", max_freq)
for center, freq in counter.items():
    weights.append(freq)
    centers.append(clustering.cluster_centers_[center])

print("Centers:" , len(centers))

hash_map = folium.Map([40.7128, -74.0060], zoom_start=11, prefer_canvas=True)
hash_map.add_child(plugins.HeatMap(map(lambda x: clustering.cluster_centers_[x], counter.elements()), radius=10))

hash_map.save(outfile="test.html")
```

## LDP Analysis on clusters

We use Apple's CMS algorithm and bin all of our NYC taxicab data via the clusters. This part of the code will take a while since the dataset is large

```python
m = 1024
k = 1024
m_prime = 1024
k_prime = 1024
epsilon_prime = 3
epsilon = 1 # 1, 0.1
threshold = 100
fragment_length = 2
max_string_length = 2
```

```python
hash_funcs = cms_helper.generate_hash_funcs(k, m)

ldp_data = []
client_cms = ClientCMS(epsilon, hash_funcs, m)

for i,element in enumerate(counter.elements()):
    if i % 50000 == 0:
        print("Stage", str(i))
    ldp_data.append(client_cms.client_cms(str(element)))

server_cms = ServerCMS(ldp_data, epsilon, k, m, hash_funcs) # Initialise the frequency oracle

print("Estimating frequencies from sketch matrix...")
keys = set(clustering.labels_)

# Generate both frequency data from the oracle and plot data to be graphed
ldp_plot_data = {}
for item in keys:
    freq = int(server_cms.freq_oracle(str(item)))
    if freq < 0:
        freq = 0
            
    ldp_plot_data[item] = freq

print("Done!")
```

This gathers the LDP frequency data and plots circles around the clusters of various length and colour weighted on the frequency estimate. Larger frequency clusters have a larger circle and display as a darker red colour on the map.

```python
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

def rgb2hex(r,g,b):
    return "#{:02x}{:02x}{:02x}".format(r,g,b)

ldp_counter = Counter(ldp_plot_data)

top_k = 100
common_points = ldp_counter.most_common(top_k)
viridis = cm.get_cmap('YlOrRd', top_k)

max_freq = common_points[0][1]
color_weights = []

for i,item in enumerate(common_points):
    color_weights.append(item[1]/max_freq)

m = folium.Map([40.7128, -74.0060], zoom_start=11, prefer_canvas=True)

for i,point in enumerate(common_points):
    center = clustering.cluster_centers_[int(point[0])]
    color = viridis(color_weights[i])
    hex_color = rgb2hex(int(color[0]*255), int(color[1]*255), int(color[2]*255))
    folium.CircleMarker([center[0], center[1]],
                        radius=point[1]/300,
                        color=hex_color,
                        fill_color=hex_color, # divvy color
                       ).add_to(m)
    
m.save(outfile="clustering1.html")
m
```

# 2) LDP Heatmap Analysis using Geohashes + Folium


The first two cells load the data and plot all of it onto a heatmap. This heatmap is saved as NYC.html

```python
data = pd.read_csv("data/train.csv")
```

```python
m = folium.Map([40.7128, -74.0060], zoom_start=11, prefer_canvas=True)

# convert to (n, 2) nd-array format for heatmap
taxi_coords = data[['pickup_latitude', 'pickup_longitude']].values
subset = taxi_coords[np.random.choice(taxi_coords.shape[0], 500000, replace=False)]

# plot heatmap
m.add_child(plugins.HeatMap(subset, radius=9))
m.save(outfile= "NYC.html")
```

These next two cells begin to hash the NYC taxi GPS coordinates using Geohashes. The geohash map (before being privatised) is saved under hash_map.html

```python
def hash_geo(geo):
    return format(np.around(geo[0], decimals=3), ".3f") + "_" + format(np.around(geo[1], decimals=3), ".3f")

hashed_geos = np.apply_along_axis(lambda x: gh.encode(x[0], x[1], precision=9), 1, subset)
counter_geo = Counter(hashed_geos)

alphabet = set()

for key in counter_geo.keys():
    for char in list(key):
        alphabet.add(char)

print("Alphabet", alphabet)
print("Number of Geohashes:", len(counter_geo))
print("Top 5 Geohashes:", counter_geo.most_common(5))
```

```python
def get_geo(hashed):
    if len(hashed.split("_")) == 1:
        return 0,0
    
    return hashed.split("_")[0] + str(np.random.randint(999)), hashed.split("_")[1] + str(np.random.randint(999))

geos = Counter(list(map(gh.decode, hashed_geos)))
print(geos.most_common(5))

hash_map = folium.Map([40.7128, -74.0060], zoom_start=11, prefer_canvas=True)
hash_map.add_child(plugins.HeatMap(geos.elements(), radius=10))

hash_map.save(outfile= "hash_map.html")
```

This is where the LDP estimation takes place. We use Apple's CMS with an epsilon of 3 to estimate frequencies based on the bins formed from Geohashes.

When plotting the data back to a heatmap, data is peturbed randomly around the cluster to imitate heat where the amount of data plotted around a geohash center is based on the frequency estimated via CMS.

```python
m = 1024
k = 1024
m_prime = 1024
k_prime = 1024
epsilon_prime = 6
epsilon = 3 # 1, 0.1
threshold = 100
fragment_length = 3
max_string_length = 9
```

```python
hash_funcs = cms_helper.generate_hash_funcs(k, m)

ldp_data = []
client_cms = ClientCMS(epsilon, hash_funcs, m)

for element in counter_geo.elements():
    ldp_data.append(client_cms.client_cms(element))

server_cms = ServerCMS(ldp_data, epsilon, k, m, hash_funcs) # Initialise the frequency oracle

print("Estimating frequencies from sketch matrix...")
hash_keys = set(hashed_geos)
print(len(hash_keys))

# Generate both frequency data from the oracle and plot data to be graphed
ldp_plot_data = {}
for i,item in enumerate(hash_keys):
    if i % 10000 == 0:
        print("Stage", str(i))
        
    freq = int(server_cms.freq_oracle(item))
    if freq < 0:
        freq = 0
               
    ldp_plot_data[item] = freq

print("Done!")
```

```python
# Plotting heatmap of LDP geo data

def add_random_geo_noise(geohash):
    geo = gh.decode(geohash)
    return geo[0]+str(np.random.randint(9999)), geo[1]+str(np.random.randint(9999))

# Convert LDP data into list of repeated elements according to frequency
ldp_freq_counter = Counter(ldp_plot_data).elements()

ldp_freq = random.sample(list(ldp_freq_counter), 500000)

# Add random geo noise to each point to generate a better heatmap
ldp_freq = list(map(add_random_geo_noise, ldp_freq))
print(len(ldp_freq))
```

The LDP reconstruction heatmap is saved under ldp_map.html

```python
ldp_map = folium.Map([40.7128, -74.0060], zoom_start=11, prefer_canvas=True)

ldp_map.add_child(plugins.HeatMap(ldp_freq, radius=9))
ldp_map.save(outfile= "ldp_map.html")
ldp_map
```
